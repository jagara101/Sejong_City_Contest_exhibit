```python
# 패키지 참조

from pandas import read_excel, DataFrame
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import sys
```

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 데이터 불러오기
origin = read_excel("../Excel Data/서울시 주행실적 통계현황_1차수정.xlsx")
origin
```

```python
origin1 = origin.copy()
```

### lineplot으로 년도에 따른 온실가스 및 주행거리 증감추세 확인하기

```python
fig, ax1 = plt.subplots(1, 2,figsize=(20,6))
ax2 = ax1[0].twinx()

# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.lineplot(data=origin1, x='대상년도', y='주행 거리', label='주행 거리', color='darkblue', ax=ax1[0])
ax1[0].legend(title='주행 거리', loc="upper right")
plt.grid()

ax3 = ax1[1].twinx()  # 두 번째 서브플롯에 대해 쌍둥이 축 생성
sb.barplot(data=origin1, x='대상년도', y='온실가스', label='온실가스', palette='Set2', ax=ax3)
ax3.legend(loc="upper left")

plt.suptitle('년도에 따른 온실가스 및 주행 거리 증감추세 ', fontweight="bold")  # 그래프 제목
plt.xlabel("년도")  # x축 제목
plt.ylabel("온실가스")  # y축 제목
plt.grid()
plt.show()  # 그래프 화면 표시
plt.close()  # 그래프 그리기 종료
```
