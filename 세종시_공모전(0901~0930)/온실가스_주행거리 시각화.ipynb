```python
# 패키지 참조

from pandas import read_excel, DataFrame
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import sys
```

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 데이터 불러오기
origin = read_excel("Excel Data/서울시 주행실적 통계현황_1차수정.xlsx")
origin
```

```python
origin1 = origin.copy()
```

### lineplot으로 년도에 따른 온실가스 및 주행거리 증감추세 확인하기

```python
fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()

# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.lineplot(data=origin1, x='대상년도', y='주행 거리',  label= '주행 거리', color='#ff6600', ax=ax1)
# plt.legend(title='주행 거리',loc= "upper right")
sb.lineplot(data=origin1, x='대상년도', y='온실가스', label='온실가스',color='#0066ff', ax=ax2)
plt.legend(loc= "upper left")

plt.title('년도에 따른 온실가스 및 주행 거리 증감추세 ', fontweight = "bold")    # 그래프 제목
plt.xlabel("년도")      # x축 제목
plt.ylabel("온실가스")   # y축 제목
plt.grid()
plt.show()               # 그래프 화면 표시
plt.close()              # 그래프 그리기 종료
```
